# Cancerous tissue detection using a CNN on lymph node samples

Data import

Based on the Kaggle dataset available at https://www.kaggle.com/competitions/histopathologic-cancer-detection 

```python
########################### Packages and seeds #################################################

import datetime
import numpy as np
import pandas as pd
import random
import os
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from sklearn.metrics import RocCurveDisplay
from PIL import Image
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
import seaborn as sn

# This sets seeds such that we get deterministic output from running the same model again
seed_num = 20
os.environ['TF_DETERMINISTIC_OPS'] = '1'
os.environ['PYTHONHASHSEED'] = str(seed_num)  # Python general
random.seed(seed_num)  # Python random
np.random.seed(seed_num)
tf.random.set_seed(seed_num)
```

```python
#################################### Loading DataFrame ########################################
main_path = "/home/apw747/prac_mach_learn/final_term_project/histopathologic-cancer-detection"
train_path = main_path + "/train"
test_path = main_path + "/test"
image_classes = np.loadtxt(main_path + "/train_labels.csv", dtype=str, delimiter=",")

df=pd.DataFrame(image_classes, columns=['id', 'label'])
df_import = shuffle(df, random_state=seed_num)

print("Initial data frame")
print(df_import.head())
print("\n")
print("Appending .tif extension to dataframe names")
df_test = df_import
df_test["id"] = df_test["id"].astype(str) + ".tif"
print(df_test.head())
```

```python
###################### Pre processing - setting an even number of both 0 and 1 and cutting down the number########

data_num = 15000
df_0 = df_test[df_test["label"] == "0"][0:data_num]
df_1 = df_test[df_test["label"] == "1"][0:data_num]

df_all = pd.concat([df_0, df_1], ignore_index=True)
df_all = shuffle(df_all, random_state=seed_num).reset_index(drop=True)
print(df_all.head())
print(df_all[["label"]].value_counts())
```

```python
######################### Train and testing - image data generator #########################################

idg = ImageDataGenerator(rescale=1. / 255.) # rescaling images
im_size = 96  # Image size is 96 X 96 pixels from dataset.
train_df, test_df = train_test_split(df_all, test_size=0.3, shuffle=False)

#Note that target size is set here because images are 96 x 96 dim, not the default.
batch_size_in = 100
train_ds = idg.flow_from_dataframe(dataframe=train_df,
                                   directory=train_path,
                                   target_size=(im_size, im_size),
                                   x_col="id",
                                   y_col="label",
                                   batch_size=batch_size_in,
                                   class_mode="binary",
                                   shuffle=False)

# Note the train_path is also used here, because we take ALL data from the train folder
test_ds = idg.flow_from_dataframe(dataframe=test_df,
                                  directory=train_path,
                                  target_size=(im_size, im_size),
                                  x_col="id",
                                  y_col="label",
                                  batch_size=batch_size_in,
                                  class_mode="binary",
                                  shuffle=False)

print("\n")
print("Shape of train_ds[0][0] is (batchsize, pixel, pixel, rgb):",train_ds[0][0].shape)
```

```python
######################################## NN for crop vs no crop ########################################

pix_size = 96 # Original pixel size
rgb_size = 3
crop_size = 25
input_shape = (pix_size, pix_size, rgb_size)

# First, we must remove all layers aside from the cropping layer, such that our network is simply:
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Cropping2D(cropping=crop_size, input_shape=input_shape))
model.summary()
#Then we call on an image from test_ds which is uncropped. Take not of the slicing of the tensor: batch size is
# involved.
```

```python
######################################## Pics for neural net diagram #################################
im_val = 14
pred_test = model.predict(test_ds[0][0])

fig = plt.figure(figsize=(5, 5))
im = test_ds[0][0][im_val]
plt.imshow(im)
plt.axis('off')
plt.show()

#Then we look at the output image. Of course, the network is not doing anything aside from cropping in this instance
# so the "pred" naming is a little misleading in this case. Either way we can print an output image after cropping.

fig = plt.figure(figsize=(5, 5))
im = pred_test[im_val]
plt.imshow(im)
plt.axis('off')
plt.show()
```

```python
################################Original vs cropped (to put in report) ############################################

im_val = 14
pred_test = model.predict(test_ds[0][0])

fig = plt.figure(figsize=(5, 5))
im = test_ds[0][0][im_val]
plt.imshow(im)
plt.title('Original')
plt.show()

#Then we look at the output image. The network is only cropping in this instance.

fig = plt.figure(figsize=(5, 5))
im = pred_test[im_val]
plt.imshow(im)
plt.title("Cropped")
plt.show()
```

```python
#######################################MODEL CREATION###################################################

pix_size = 96 # Original pixel size
rgb_size = 3

n_epochs = 20
DropoutValue = 0.875
learning_rate = 0.0002

crop_size = 25
input_shape = (pix_size, pix_size, rgb_size)

model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Cropping2D(cropping=crop_size, input_shape=input_shape))

model.add(keras.layers.Conv2D(32, (3, 3), activation='relu'))
model.add(keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Dropout(DropoutValue))

model.add(keras.layers.Conv2D(64, (5, 5), activation='relu'))
model.add(keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Dropout(DropoutValue))

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(256, activation='relu'))
model.add(tf.keras.layers.Dropout(DropoutValue))
model.add(keras.layers.Dense(1, activation='sigmoid'))

#Binary cross_entropy set as the accuracy metric has a default threshold value of 0.5
loss_fn = "binary_crossentropy"
optimizer = keras.optimizers.Adam(learning_rate=learning_rate)
model.compile(optimizer=optimizer, loss=loss_fn, metrics=['accuracy'])
model.summary()
```

```python
#####################GENERATE DATA AND RUN NETWORK#############################
history = model.fit(train_ds, validation_data=test_ds, batch_size=batch_size_in, epochs=n_epochs,
                    callbacks=[tf.keras.callbacks.EarlyStopping(monitor="val_loss", patience=8)])
model.summary()
```

```python
################################### Predicted ################################################
pred = model.predict(test_ds)
print(pred[0])
```

```python
################################### Predicted vs True plot ################################################

# Tumour tissue present = 1
# Tumour tissue not present = 0

fig = plt.figure(figsize=(12, 9))
for i in range(0,20):
    ax = fig.add_subplot(4, 5, i+1, xticks=[], yticks=[])
    im = test_ds[0][0][i]
    plt.imshow(im)
    #Note that we apply the standard threshold value when comparing true and predicted values here.
    lab_true = test_ds[0][1][i]
    if pred[i] >= 0.5:
        lab_pred = 1
    else:
        lab_pred = 0
    ax.set_title('T: '+ str(int(lab_true)) + ","+ " P: "+ str(lab_pred))
plt.show()
```

```python
################### Training and test accuracy as a function of training epochs ##########################

keys = history.history.keys()
print("\033[1mTraining history contains the following keys:\033[0m")
for key in keys:
    print("\t", key)

print("\n\033[1mDisplay the evolution of the accuracy as a function of the training epoch\033[0m\n")
print("  N(Epochs)           = ", n_epochs)
print("  accuracy (train)     = ", history.history['accuracy'])
print("  accuracy (validate)  = ", history.history['val_accuracy'])
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validate'], loc='lower right')
plt.show()
plt.clf()

print("\n\033[1mDisplay the evolution of the loss as a function of the training epoch\033[0m\n")
print("  N(Epochs)        = ", n_epochs)
print("  loss (train)     = ", history.history['loss'])
print("  loss (validate)  = ", history.history['val_loss'])

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validate'], loc='upper right')
plt.show()
```

```python
########################################### Creating ROC curve ########################################

target_list = []
for bnum in range(0, len(test_ds)):
    for a in range(0, len(test_ds[bnum][1])):
        target_list.append(int(test_ds[bnum][1][a]))
#print(target_list)
print(len(pred), len(target_list))

RocCurveDisplay.from_predictions(np.array(target_list), np.array(pred))
plt.show()
```
